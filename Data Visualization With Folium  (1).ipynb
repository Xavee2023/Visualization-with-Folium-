```python
# import libraries

import pandas as pd
import numpy as np
import folium
```

```python
# Read dataset into pandas

filepath = r"C:\Users\ejiro\Downloads\Police_Department_Incidents_-_Previous_Year__2016_.csv"
df = pd.read_csv(filepath, encoding='ISO-8859-1')
```

```python
df.head()
```

```python
# count the unique catefgory of crime values
df["Category"].value_counts()
```

```python
# select only the larceny/theft crimes as it is the most common

# create a boolean mask to select rows with violent crimes
df_crimes = df["Category"] == "LARCENY/THEFT"
```

```python
# apply the mask to the dataframe
crimes = df[df_crimes]
```

```python
crimes.head()
```

```python
# confirm only larceny/theft values are in the dataset now
crimes['Category'].value_counts()
```

```python
# Get the first 100 larceny/theft crimes in the dataset 
limit = 100
crimes = crimes.iloc[0:limit, :]
```

```python
crimes.shape
```

```python
# San francisco longitute and latitude
latitude = 37.77
longitude = -122.42
```

```python
# import Figure class for resizing maps and pass the desired width and height in pixels:
from branca.element import Figure
fig=Figure(width=550,height=350)
```

```python
# create map and display it

sf_map = folium.Map(location=[latitude, longitude], zoom_start=12)

sf_map
```

```python
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(crimes.Y, crimes.X):
    incidents.add_child(
        folium.vector_layers.CircleMarker(
           [lat, lng],
           radius = 7,
           color='yellow',
           fill=True,
           fill_color='blue', 
           fill_opacity=0.6
      )
        
    )
    
 # add pop-up to each marker on the map
latitudes = list(crimes.Y)
longitudes = list(crimes.X)
labels = list(crimes.Category)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(sf_map)
    
    
    
# add incidents to the map
sf_map.add_child(incidents)
```
